# Data Cleaning part for airbnb booking analysis (Chicago and New Orleans)
**By Pihu Goyal**

**Objective:**  
The goal of this data cleaning process was to prepare the Airbnb datasets for **accurate analysis and visualization** by handling missing values, ensuring consistency, and merging both datasets into a single structured table.  

**Steps Taken:**  

**1. Fixing Column Names**  
- Renamed columns where necessary to maintain **consistent naming conventions** across both datasets.  
- Ensured column names **do not have spaces or special characters** that might cause issues in analysis.  

**2. Dropping Unnecessary Columns**  
- Removed columns that were **not relevant to our analysis**, such as URLs, image links, and redundant calculated fields.  
- This helped **reduce dataset size and improve efficiency** in processing.  

**3. Checking & Removing Duplicates**  
- Verified that no duplicate records were present in both datasets.  
- Checked for duplicate `id` values and ensured uniqueness.  

**4. Converting Data Types**  
- Converted numerical columns like `price`, `host_acceptance_rate`, and `reviews_per_month` to **appropriate data types** (float/int).  
- Converted date-related fields (`host_since`, `first_review`, `last_review`) to **datetime format** for time-series analysis.  
- Converted boolean-like columns (`has_availability`, `host_is_superhost`) to **Boolean (True/False)** format.  

**5. Handling Missing Values**  
- **For `first_review`, `last_review`, `review_scores_value`, `reviews_per_month`:**  
  - These were missing for listings **with zero reviews**, so we filled them with `0` or `NaT` (for date fields).  
- **For `host_acceptance_rate`:**  
  - Found that when `host_acceptance_rate` was missing, the host was **not a superhost**, so we filled it with a lower median value.  
- **For `price`:**  
  - Imputed missing values based on **median price grouped by `neighbourhood_cleansed` and `room_type`** for better accuracy.  
- **For `host_neighbourhood`:**  
  - Filled missing values using the most frequent `neighbourhood_cleansed` value for that listing's area.  
- **For `has_availability`:**  
  - Converted missing values to **False (`f`)**, assuming no availability was reported.  

**6. Adding City Labels & Merging Data**  
- Added a **"City"** column to differentiate listings from **Chicago** and **New Orleans**.  
- Merged both datasets into a **single table** using `pd.concat()`.  

**7. Exporting Cleaned Data**  
- Exported the final cleaned dataset as `"Cleaned_Airbnb_Data.csv"` for **further analysis and visualizations**.  

**Conclusion:**  
This cleaning process ensures that the Airbnb dataset is **ready for accurate insights and visual storytelling**. With properly formatted data, we can now explore **trends in pricing, host behavior, and neighborhood popularity** effectively!

# Changing the data types for the required columns


```python
# Convert 'id' and 'host_id' to string (IDs should not be treated as numbers)
df_chicago["id"] = df_chicago["id"].astype(str)
df_chicago["host_id"] = df_chicago["host_id"].astype(str)

# Convert 'price' to numeric (after removing any currency symbols)
df_chicago["price"] = df_chicago["price"].replace('[\$,]', '', regex=True).astype(float)

# Convert boolean-like columns to proper boolean type
df_chicago["host_is_superhost"] = df_chicago["host_is_superhost"].map({"t": True, "f": False}).astype(bool)
df_chicago["host_identity_verified"] = df_chicago["host_identity_verified"].map({"t": True, "f": False}).astype(bool)
df_chicago["instant_bookable"] = df_chicago["instant_bookable"].map({"t": True, "f": False}).astype(bool)
df_chicago["has_availability"] = df_chicago["has_availability"].map({"t": True, "f": False}).astype(bool)


# Convert 'host_since', 'first_review', 'last_review' to datetime
df_chicago["host_since"] = pd.to_datetime(df_chicago["host_since"])
df_chicago["first_review"] = pd.to_datetime(df_chicago["first_review"])
df_chicago["last_review"] = pd.to_datetime(df_chicago["last_review"])

# Convert 'minimum_nights', 'maximum_nights', 'availability_365', 'number_of_reviews' to integer
int_cols = ["minimum_nights", "maximum_nights", "availability_365", "number_of_reviews"]
df_chicago[int_cols] = df_chicago[int_cols].astype(int)

# Convert 'review_scores_value' and 'reviews_per_month' to float
df_chicago["review_scores_value"] = df_chicago["review_scores_value"].astype(float)
df_chicago["reviews_per_month"] = df_chicago["reviews_per_month"].astype(float)

# Remove the '%' symbol and convert to float
df_chicago["host_acceptance_rate"] = df_chicago["host_acceptance_rate"].str.replace('%', '').astype(float)
```

```python
df_chicago.dtypes
```

# Handling the missing values

```python
missing_values = df_chicago.isnull().sum()
missing_percent = (missing_values / len(df_chicago)) * 100
missing_df = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percent})
print(missing_df[missing_df["Missing Values"] > 0].sort_values(by="Percentage", ascending=False))
```

```python
# Why Keeping NaT is Fine?
# Safe for Time-Series Analysis → Pandas functions ignore NaT automatically in calculations.
# Avoids Incorrect Assumptions → Filling missing dates with 0 or "Unknown" would break datetime operations.
# Filters & Aggregations Work → You can still filter, sort, or group by dates without issues.
df_chicago["first_review"] = df_chicago["first_review"].fillna(pd.NaT) # Keeps it as NaT (Not a Time)
df_chicago["last_review"] = df_chicago["last_review"].fillna(pd.NaT)
```

```python
# Fill missing values with 0
df_chicago["review_scores_value"] = df_chicago["review_scores_value"].fillna(0)
df_chicago["reviews_per_month"] = df_chicago["reviews_per_month"].fillna(0)
```

```python
# Fill missing values with "f" (False)
df_chicago["has_availability"] = df_chicago["has_availability"].fillna(False)
```

```python
# Fill missing prices using the median price of the same 'neighbourhood_cleansed' and 'room_type'
df_chicago["price"] = df_chicago.groupby(["neighbourhood_cleansed", "room_type"])["price"].transform(lambda x: x.fillna(x.median()))
df_chicago["price"] = df_chicago["price"].fillna(df_chicago["price"].median())
```

```python
# Fill missing host_location values with "Chicago, USA"
# Assumption: If host location is missing, they are likely in the same city as the listing
df_chicago["host_location"] = df_chicago["host_location"].fillna("Chicago, IL")
```

```python
# Compute a lower percentile (25th percentile) for a more reasonable fill
lower_percentile = df_chicago["host_acceptance_rate"].quantile(0.25)  # 25th percentile

# Fill missing values with the lower percentile where host_is_superhost is False
df_chicago["host_acceptance_rate"] = df_chicago["host_acceptance_rate"].fillna(lower_percentile)
```

```python
# Fill missing values based on host_acceptance_rate
df_chicago.loc[(df_chicago["host_is_superhost"].isna()) & (df_chicago["host_acceptance_rate"] >= 90), "host_is_superhost"] = True
df_chicago.loc[(df_chicago["host_is_superhost"].isna()) & (df_chicago["host_acceptance_rate"] < 90), "host_is_superhost"] = False
```

```python
# Fill missing 'host_neighbourhood' based on the most common value (mode) of 'neighbourhood_cleansed'
df_chicago["host_neighbourhood"] = df_chicago.groupby("neighbourhood_cleansed")["host_neighbourhood"].transform(lambda x: x.fillna(x.mode()[0] if not x.mode().empty else "Unknown"))
```

```python
df_chicago['host_name'] = df_chicago['host_name'].fillna('Unknown')
```

```python
# Convert to datetime if not already
df_chicago['host_since'] = pd.to_datetime(df_chicago['host_since'], errors='coerce')

# Find the latest (most recent) date in the column
latest_date = df_chicago['host_since'].max()

# Replace missing values with the latest date
df_chicago['host_since'] = df_chicago['host_since'].fillna(latest_date)
```

```python
# Calculate the mean (excluding missing values)
mean_value = df_chicago['host_listings_count'].mean()

# Replace missing values with the mean
df_chicago['host_listings_count'] = df_chicago['host_listings_count'].fillna(mean_value)
```
